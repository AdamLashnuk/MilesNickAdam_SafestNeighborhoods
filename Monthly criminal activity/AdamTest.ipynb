This notebook is part of our final project on choosing the best neighborhood in Pittsburgh. There are many things people might care about, like price, convenience, or entertainment, but after talking with my group, we decided that safety matters the most. Safety covers things like crime levels and emergency activity. My part of the project focuses on the number of criminal incidents in each neighborhood.

Our main metric is safety, and we use different datasets to measure it. I looked specifically at the “Monthly Criminal Activity” data to see how many crimes happen in each neighborhood. Fewer crimes usually means a safer place, so the neighborhoods with the lowest counts should be the strongest candidates.

```python
import pandas as pd
import matplotlib.pyplot as plt

data_set = pd.read_excel("MonthlyCriminalActivity.xlsx")
column = "Neighborhood"

# takes dataset, then column in dataset, then counts how many appear
crimeCounts = data_set[column].value_counts().tail(10)

# plotting
plt.title("Criminal activity")
plt.xlabel("Zone")
plt.ylabel("Count")

plt.style.use('fivethirtyeight')
crimeCounts.plot(kind='bar')
plt.show()
```

Based on the bar chart, the five neighborhoods with the lowest crime counts are:
Saint Clair
Ridgemont
Chartiers
New Homestead
Swisshelm Park
Saint Clair has the fewest incidents, making it the safest according to this dataset. Ridgemont and Chartiers also show very low numbers, and New Homestead and Swisshelm Park follow closely behind.

Conclusion
Looking only at crime data, Saint Clair stands out as the safest neighborhood. Ridgemont and Chartiers are close behind and could also be considered very safe.
So, based on this sub-metric alone, Saint Clair is my choice for the safest neighborhood in Pittsburgh. Once we combine this with the other group metrics, we’ll be able to make a final decision on the overall best neighborhood.
