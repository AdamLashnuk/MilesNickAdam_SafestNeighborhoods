**Data Set 1: Monthly Criminal Activity (Adam)**

Our main metric is safety, and we use different datasets to measure it. I looked specifically at the “Monthly Criminal Activity” data to see how many crimes happen in each neighborhood. Fewer crimes usually means a safer place, so the neighborhoods with the lowest counts should be the strongest candidates.

```python
import pandas as pd
import matplotlib.pyplot as plt

data_set = pd.read_excel("MonthlyCriminalActivity.xlsx")
column = "Neighborhood"

# takes dataset, then column in dataset, then counts how many appear
crimeCounts = data_set[column].value_counts().tail(10)

# plotting
plt.title("Criminal activity")
plt.xlabel("Zone")
plt.ylabel("Count")

plt.style.use('fivethirtyeight')
crimeCounts.plot(kind='bar')
plt.show()
```

Based on the bar chart, the five neighborhoods with the lowest crime counts are: Saint Clair Ridgemont Chartiers New Homestead Swisshelm Park Saint Clair has the fewest incidents, making it the safest according to this dataset. Ridgemont and Chartiers also show very low numbers, and New Homestead and Swisshelm Park follow closely behind.

**Conclusion:**
Looking only at crime data, Saint Clair stands out as the safest neighborhood. Ridgemont and Chartiers are close behind and could also be considered very safe.
So, based on this sub-metric alone, Saint Clair is my choice for the safest neighborhood in Pittsburgh. Once we combine this with the other group metrics, we’ll be able to make a final decision on the overall best neighborhood.

**Data Set 2: Police Arrest Data (Nick)**

**Introduction** - For our project on finding the best neighborhood in Pittsburgh, I focused on looking at safety using the Pittsburgh Police Arrest dataset. Since the dataset includes every recorded arrest along with the neighborhood where it happened, it gave me a good way to compare different areas. By counting how often each neighborhood showed up and turning that into a bar graph, I was able to see which places had higher or lower levels of police activity. This helped me get a better sense of how safety might differ across the city and which neighborhoods could be considered safer based on arrest frequency.

```python
import pandas as pd
import matplotlib.pyplot as plt

data_set = pd.read_csv("Pittsburgh_Police_Arrest_Data.csv")

#Looking at the column I want from the csv file
col = 'INCIDENTNEIGHBORHOOD'

#Getting the least amount of arrests in the neighborhoods to see which one is the safest
arrest_counts = data_set[col].value_counts().tail(10)

#Creating the Bar Graph
arrest_counts.plot.bar(color='blue')
plt.title("Safest Neighborhoods in Pittsburgh")
plt.xlabel("Arrest Location (Neighborhood)")
plt.ylabel("Frequency (Amount of Arrests)")
plt.xticks(rotation=45, ha='right')

plt.show()
```

**Submetric conclusion** - Using the Pittsburgh Police Arrest dataset made it easy to see which neighborhoods show up the least when it comes to arrests. By counting how often each neighborhood appeared and graphing it, I was able to get a clearer picture of which area might be considered “safer” based on lower arrest numbers. It’s definitely not a perfect measure of safety, but the dataset gives a solid starting point.
